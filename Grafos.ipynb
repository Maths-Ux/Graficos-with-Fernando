```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("frutas.CSV")
df
```

```python
Vendas_por_Fruta = df.groupby("Nome Produto")["Qtd"].sum()
Vendas_por_Fruta
plt.bar(Vendas_por_Fruta.index, height=Vendas_por_Fruta.values, color=["goldenrod","tomato","darkslateblue"])
plt.title("Quantidade Total de Frutas")
plt.xlabel("Frutas", y=-0.1)
plt.ylabel("Quantidade")
plt.xticks(rotation=45)
plt.grid(axis="y", alpha=1)
```

```python
Lucro_por_fruta = df.groupby("Nome Produto")["Lucro Total"].sum()
Vendas_por_Fruta
plt.pie(Lucro_por_fruta.values, labels = Lucro_por_fruta.index, colors=["goldenrod","tomato","darkslateblue"], autopct='%1.1f%%')
plt.show
```

```python
mapa_calor = df[["Custo", "Valor", "Qtd", "Lucro Total"]]

correlacao  = mapa_calor.corr()
plt.imshow(correlacao, cmap="coolwarm", interpolation="nearest")
plt.colorbar(label = "Correlação")
plt.xticks(range(len(correlacao.columns)), correlacao.columns)
plt.yticks(range(len(correlacao.columns)), correlacao.columns)
```
